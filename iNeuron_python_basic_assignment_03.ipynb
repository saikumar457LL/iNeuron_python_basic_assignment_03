9.How do you make a function variable refer to the global variable?

Ans:

    i will use global keyword befer refering the function variable to a variable which is outside of a function
    
    Eg: follow the below code

```python
some_var = 8

def some_func():
    global some_var
    local_var = some_var
    return local_var
    
print(some_func()) # prints 8

# another

def some_fun():
    global k # making k as global
    k = 3
    return k

print(some_fun()) # prints 3
print(k) # prints 3
```
